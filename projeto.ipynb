```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv('brasil.csv').filter(items=['ano', 'emissao'])
```

```python
df = df.groupby(by='ano').sum().reset_index()
```

```python
df
```

```python
anos = df['ano']
emissoes = df['emissao']
```

```python
emissoes
```

```python
plt.bar(anos, emissoes)
plt.title("Emissão de carbono")
plt.xlabel('Ano')
plt.ylabel('Emissão')
plt.show()
```
